```python
import matplotlib.pyplot as plt
```

```python
x = [32, 64, 128, 256, 512, 1024, 2560, 5120, 7680, 8960, 10240]
y = [1.982, 1.165, 0.678, 0.468, 0.417, 0.34, 0.224, 0.231, 0.202, 0.489, 0.963]
```

```python
# plt.figure(figsize=(10, 10))
plt.plot(x, y, marker='o', color='r')
plt.xlabel('Number of threads')
plt.ylabel('Minimum time')
plt.savefig('threads_vs_time', quality=95)
```
